```python
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
client = MongoClient()
db = client.work
aссural = db.accural
payments = db.payments

def test(accrual, payment):
    accrual = list(aссural.aggregate([{'$sort': {'date': -1}}]))
    payment = list(payments.aggregate([{'$sort': {'date': 1}}]))
    true_list = []
    false_list = []

    for p in payment:
        for a in accrual:
            if p['data'] > a['data']:
                if p['month'] == a['accuralmonth']:
                    p['accrual'] = a
                    true_list.append(p)
                    accrual.remove(a)
                else:
                    p['accrual'] = accrual[-1]
                    true_list.append(p)
                    accrual.pop()
                break
    if len(payment) > len(true_list):
        while len(payment) > len(true_list):
            false_list.append(payment[-1])
            payment.remove(payment[-1])

    print('Соответствия: ')
    pprint(true_list)
    print('Платежи без долга: ')
    pprint(false_list)
test(aссural, payments)
```
